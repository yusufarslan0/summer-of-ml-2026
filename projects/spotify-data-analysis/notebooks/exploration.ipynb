### Duplicate Track ID's

```python
repeated_track_rows = df[df["track_id"].duplicated(keep=False)]
print(len(repeated_track_rows)) # There are 40900 duplicate row group 
```

```python
duplicated_id_rows = df[df["track_id"].duplicated(keep="first")]
print(len(duplicated_id_rows)) # There are 24259 first occurrences of the duplicate track IDs
```

```python
print(df.duplicated(keep=False).sum()) # 894 rows belong to a completely duplicate group.
```

```python
print(df.duplicated(keep="first").sum()) # 450 rows are duplicate copies beyond the first occurrence.
```

```python
repeated_track_rows.head(10) # I wanted to inspect the duplicates individually to understand reasons behind the duplicate rows
```

```python
for duplicate_id in repeated_track_rows.head(10)["track_id"]:
    display(df[df["track_id"] == duplicate_id]) 
```

```python
non_identical_number = 0
special_ids = []

df_genre_dropped = df.drop(columns=["track_genre"])

for repeated_id in repeated_track_rows["track_id"].head(1000):
    id_group = df_genre_dropped[df["track_id"] == repeated_id]

    is_all_identical = all(id_group.duplicated(keep=False))

    if not is_all_identical:
        non_identical_number += 1
        special_ids.append(repeated_id)

print(non_identical_number)
```

Initial inspection of several duplicated track_id values suggested that repeated tracks differed only in the track_genre column. To verify this observation, the first 1,000 duplicated track_id groups were examined programmatically. 

Among the first 1000 duplicated track_id groups inspected programmatically, only 28 groups contained differences beyond `track_genre`.


```python
for special_id in special_ids:
    display(df[df["track_id"] == special_id ])
```

Among the first 1,000 duplicated `track_id` groups inspected, the only observed differences outside `track_genre` were in the `popularity` column.

One possible explanation is that the dataset creator queried the Spotify API at different times. Since Spotify's popularity metric changes over time, repeated observations of the same track may have different popularity values while the audio features remain unchanged. This explanation is plausible but cannot be confirmed from the available dataset documentation.

```python
non_identical_number = 0

df_genre_pop_dropped = df.drop(columns=["track_genre", "popularity"])
for special_id in special_ids:
    id_group = df_genre_pop_dropped[df["track_id"] == special_id] # rows for a specific duplicated ID 
    is_all_identical = all(id_group.duplicated(keep=False)) # See if the rows w/o genre identical
    if is_all_identical == False:
        non_identical_number += 1
print(non_identical_number)
```
